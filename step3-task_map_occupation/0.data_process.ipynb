```python
import os
import pandas as pd
dataset_name = 'en'
all_data_task_topk = pd.read_csv('../step2-task_map_occupation/outputs/en/all_data_task_topk.csv', sep='\t')
all_data_task_topk['task_candidate'] = all_data_task_topk.apply(lambda x: [x[f"task{i}"] for i in range(10)], axis=1)
all_data_task_topk = all_data_task_topk[['occupation', 'task', 'task_candidate']]
all_data_task_topk.to_parquet(f'data/{dataset_name}/task_topkoccupation.parquet')
all_data_task_topk = all_data_task_topk.reset_index()
all_data_task_topk.columns = ['index', 'occupation', 'job description', 'occupation_candidate']
all_data_task_topk = all_data_task_topk[['index', 'job description', 'occupation_candidate']]
for seed in range(5):
    if not os.path.exists(f'data/{dataset_name}/seed{seed}/input0'):
        os.makedirs(f'data/{dataset_name}/seed{seed}/input0')
    all_data_task_topk.to_parquet(f'data/{dataset_name}/seed{seed}/input0/task_candidate.parquet')
all_data_task_topk
```
