```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
x = [0, 1, 2, 3, 4, 5]
y1_random = [55, 56, 55, 56, 59, 57]
y2_random = [54, 57, 56, 55, 56, 55]

y1_entropy = [55, 55, 56, 57, 56, 58]
y2_entropy = [54, 54, 54, 55, 54, 55]

y1_cal = [55, 55, 56, 57, 58, 60]
y2_cal = [54, 55, 55, 56, 57, 58]

# 初始化画布
fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

# Plot for Random
axs[0].plot(x, y1_random, 'o-', label='COTR-SRM(Random)', color='#B86D42', markersize=8)  
axs[0].plot(x, y2_random, '^-', label='COTR-SRM w/o pairwise loss', color='#46835D', markersize=8)  
axs[0].set_title('COTR-SRM(Random)')
axs[0].set_ylim([54, 60])

# Plot for Entropy
axs[1].plot(x, y1_entropy, 'o-', label='COTR-SRM(Entropy)', color='#B86D42', markersize=8)  
axs[1].plot(x, y2_entropy, '^-', label='COTR-SRM w/o pairwise loss', color='#46835D', markersize=8)  
axs[1].set_title('COTR-SRM(Entropy)')
axs[1].set_ylim([54, 60])

# Plot for CAL
axs[2].plot(x, y1_cal, 'o-', label='COTR-SRM(CAL)', color='#B86D42', markersize=8)  
axs[2].plot(x, y2_cal, '^-', label='COTR-SRM w/o pairwise loss', color='#46835D', markersize=8)  
axs[2].set_title('COTR-SRM(CAL)')
axs[2].set_ylim([54, 60])

# 统一配置
for ax in axs:
    ax.legend(loc='upper left')
    ax.grid(axis='y',  linestyle='--') 
    ax.set_xticks(x)
    ax.set_yticks(np.arange(54, 61, 1))

# 显示图表
plt.tight_layout()
plt.show()
```
