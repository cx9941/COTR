```python
import pandas as pd
task_label = pd.read_excel('DWA Reference.xlsx')
task_label = task_label[['Element Name', 'IWA Title', 'DWA Title']]
task_label.to_csv('../data/task.csv', index=None,sep='\t')
task_label
```

```python
job2task = pd.read_excel('Tasks to DWAs.xlsx')[['Title', 'Task', 'DWA Title']]
job2task = pd.merge(job2task, task_label, on=['DWA Title'])
job2task.columns = ['job_title', 'job_description', 'DWA Title', 'Element Name', 'IWA Title']
job2task.to_csv('../data/task_data.csv', index=None,sep='\t')
job2task
```

```python
job2task = job2task.sample(frac=1).reset_index(drop=True)
job2task
```

```python
train_data = job2task.groupby(['Element Name', 'IWA Title', 'DWA Title']).agg(list)[['job_description']]
train_data['job_description'] = train_data['job_description'].apply(lambda x: x[: max(int(len(x)*0.8), 1)])
train_data = train_data.reset_index().explode('job_description')
train_data = job2task[job2task['job_description'].isin(train_data['job_description'].tolist())]
train_data.to_csv('../data/train.csv', index=None, sep='\t')
train_data
```

```python
eval_data = job2task[~job2task['job_description'].isin(train_data['job_description'].tolist())]
eval_data.to_csv('../data/eval.csv', index=None, sep='\t')
eval_data
```
