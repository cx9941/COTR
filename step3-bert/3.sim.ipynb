```python
import torch
dataset_name = 'en'
occupation_embedding = torch.load(f'outputs/{dataset_name}/occupation_embedding.pt')
title_embedding = torch.load(f'outputs/{dataset_name}/title_embedding.pt')
```

```python
import pandas as pd
import os
occupation = pd.read_csv(f'data/{dataset_name}/occupation.csv', sep='\t')
title = pd.read_csv(f'data/{dataset_name}/title.csv', sep='\t')
```

```python
from utils import cosine_similarity_matrix
similarity_matrix = cosine_similarity_matrix(title_embedding, occupation_embedding)
topk_indices = torch.topk(similarity_matrix, dim=-1,k=10).indices
topk_value = torch.topk(similarity_matrix, dim=-1,k=10).values
```

```python
import pandas as pd
topk_task = pd.DataFrame(topk_indices.numpy())
topk_values = pd.DataFrame(topk_value.numpy())
```

```python
for col in topk_task.columns:
    title[f"occupation{col}"] = topk_task[col].apply(lambda x: occupation['occupation'][x]) 
    title[f"occupation{col}_sim"] = topk_values[col] 
title.to_csv(f'outputs/{dataset_name}/title_top10_occupation.csv', index=None, sep='\t')
title.head()
```
