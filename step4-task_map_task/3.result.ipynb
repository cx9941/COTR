```python
import pandas as pd
import os

dataset_name = 'en'
seed=0
result_dir = f'results/{dataset_name}/seed{seed}'
final_ans = []
task_topkoccupation = pd.read_parquet('/data/chenxi/research/conference/TMIS/step3-task_map_occupation/data/en/task_topkoccupation.parquet')[['occupation', 'task', 'task_sim']]
task_topkoccupation.columns = ['origin_occupation', 'job description', 'sim']
task_candidate = pd.read_parquet(f'data/{dataset_name}/task_candidate.parquet')[['occupation', 'job description']]
task_topkoccupation = pd.merge(task_topkoccupation, task_candidate, on=['job description'])
task_topkoccupation
```

```python
final_ans = []
for i in range(len(os.listdir(result_dir))):
    df = pd.read_parquet(f"{result_dir}/result{i}.parquet")
    final_ans.append(df)
final_ans = pd.concat(final_ans)[['job description', 'task']].drop_duplicates('job description')
final_ans = pd.merge(final_ans, task_topkoccupation, on=['job description']).sort_values(['sim'], ascending=[False])
final_ans = final_ans[['origin_occupation', 'job description', 'occupation', 'task', 'sim']]
final_ans
```

```python
topk = 100
final_ans[:topk].to_excel(f'results/{dataset_name}/seed{seed}_top{topk}.xlsx', index=None)
final_ans
```
