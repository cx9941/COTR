```python
import pandas as pd
import os

dataset_name = 'en'
seed = 1
df = pd.read_parquet(f'../step3-task_map_occupation/results/{dataset_name}/seed{seed}.parquet')
task_occupation = pd.read_csv(f'../data/{dataset_name}/task.csv', sep='\t')
task_occupation_group = task_occupation.groupby('occupation').agg(list)
df = pd.merge(df, task_occupation_group, on='occupation')
df.to_parquet(f'data/{dataset_name}/task_candidate.parquet')
df = df[['job description', 'task']]
df.columns = ['job description', 'task_candidate']
for seed in range(5):
    if not os.path.exists(f'data/{dataset_name}/seed{seed}/input0'):
        os.makedirs(f'data/{dataset_name}/seed{seed}/input0')
    df.reset_index().to_parquet(f'data/{dataset_name}/seed{seed}/input0/task_candidate.parquet')
df
```

```python
seed = 1
seed1_top100 = pd.read_excel(f'/data/chenxi/research/conference/TMIS/step3-task_map_occupation/results/{dataset_name}/seed{seed}_top100.xlsx')
df = pd.merge(df, seed1_top100, on=['job description'])[:50]
df.to_excel(f'data/{dataset_name}/seed{seed}_top50.xlsx', index=None)
```
