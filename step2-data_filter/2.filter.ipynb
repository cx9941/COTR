```python
import pandas as pd
ans = {}
for seed in range(12):
    ans[seed] = []
    for i in range(3517):
        text = open(f'outputs/seed{seed}/{i}.txt', 'r').read()
        text = text[text.find('返回“是”或“否”\n<|im_end|>\n<|im_start|>') + len('返回“是”或“否”\n<|im_end|>\n<|im_start|>'):]
        ans[seed].append(text)
ans = pd.DataFrame(ans)
ans
```

```python
for seed in ans:
    ans[seed] = ans[seed].apply(lambda x: int('否' not in x))
ans['is_task'] = ans.apply(lambda x: sum([x[i] for i in ans.columns]), axis=1)
ans = ans[['is_task']]
job_re_description_ch = pd.read_csv('data/jp-job_re_description_ch.csv', sep='\t')
job_re_description_ch = pd.concat([job_re_description_ch, ans], axis=1)
job_re_description_ch['annotated'] = job_re_description_ch.apply(lambda x: int((x['is_task']>5) or (x['is_task']>=5 and x['length']>=40)), axis=1)
job_re_description_ch.to_excel('results/jp-job_re_description_ch_filter.xlsx', index=None)
job_re_description_ch
```

```python
job_re_description_ch = pd.read_excel('results/jp-job_re_description_ch_filter1.xlsx')
job_re_description = pd.read_excel('data/jp-job_re_description_ch.xlsx')
job_re_description.columns = ['job_title', 'job_description', 'length']
job_re_description = pd.merge(job_re_description, job_re_description_ch, on=job_re_description.columns.tolist(), how='left')
job_re_description = job_re_description.fillna(0).reset_index()
job_re_description.to_excel('results/jp-job_re_description_ch_annotated.xlsx', index=None)
job_re_description
```
