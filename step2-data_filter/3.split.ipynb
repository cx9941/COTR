```python
import pandas as pd
df = pd.read_excel('results/jp-job_re_description_ch_annotated1.xlsx')
df.columns = ['index', 'job_title', 'task', 'task_len', 'annotated']
df
```

```python
job_re_description = pd.read_excel('data/jp-job_re_description.xlsx')
job_re_description['annotated'] = df['annotated']
job_re_description
```

```python
import re
job_re_description['task'] = job_re_description.apply(lambda x: re.split(r'・|-|●|•|～|◎|(\d)|\d\.|\d、|．|。', x['task']), axis=1)
job_re_description = job_re_description.explode('task')
job_re_description = job_re_description[~job_re_description['task'].isna()]
job_re_description['task_len'] = job_re_description['task'].apply(len)
job_re_description = job_re_description.sort_values('task_len', ascending=False)
job_re_description = job_re_description[(job_re_description['annotated']==1) & (job_re_description['task_len']>5)].drop('annotated', axis=1).drop_duplicates('task')
job_re_description.to_excel('results/jp-job_re_description.xlsx', index=None)
job_re_description
```
