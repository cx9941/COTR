```python
import pandas as pd
from utils import gen

dataset_name = 'jp'
final_ans = pd.read_csv(f'../step4-llm_select/results/{dataset_name}/{dataset_name}_gpt-llm-bert.csv')
final_ans.to_csv(f'data/{dataset_name}.csv', index=None)
final_ans
```

```python
final_ans = pd.read_csv(f'data/{dataset_name}.csv')
final_ans['bert_task'] = final_ans['bert_task'].apply(eval)
final_ans = final_ans[['job_title', 'job_description', 'bert_task']]
for i in range(100):
    final_ans[f"task{i}"] = final_ans['bert_task'].apply(lambda x: x[i])
final_ans = final_ans.drop('bert_task', axis=1)
final_ans['prompt'] = final_ans.apply(lambda x: gen(x, dataset_name), axis=1)
final_ans
```

```python
import os
if not os.path.exists(f"data/{dataset_name}/inputs/"):
    os.makedirs(f"data/{dataset_name}/inputs/")
for i in range(len(final_ans)):
    with open(f"data/{dataset_name}/inputs/{i}.txt", 'w') as w:
        w.write(final_ans['prompt'][i])
```
