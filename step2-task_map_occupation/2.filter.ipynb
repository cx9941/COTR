```python
import pandas as pd
dataset_name = 'en'
df = pd.read_csv(f'outputs/{dataset_name}/job_task.csv', sep='\t')
df = df[['title', 'task']]
df.columns = ['title', 'description']
df
```

```python
import pandas as pd
df = pd.read_csv(f'outputs/{dataset_name}/title_task.csv', sep='\t')
df['length'] = df['description'].apply(lambda x: len(x.split(' ')))
df = df[(df['length']>10)&(df['length']<50)]
df = df.sort_values('length', ascending=False)
df
```

```python
import re
def filter(text):
    match = re.findall(r'salary', text)
    return len(match)>0
df = df[~df['description'].apply(filter)]
df.to_csv(f'outputs/{dataset_name}/title_task.csv', index=None, sep='\t')
df
```
