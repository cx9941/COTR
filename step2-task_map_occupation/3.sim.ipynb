```python
import torch
dataset_name = 'en'
task_embedding = torch.load(f'outputs/{dataset_name}/task_embedding.pt')
all_data_embedding = torch.load(f'outputs/{dataset_name}/all_data_embedding.pt')
all_data_task_embedding = torch.load(f'outputs/{dataset_name}/all_data_task_embedding.pt')
task_occupation_embedding = torch.load(f'outputs/{dataset_name}/task_occupation_embedding.pt')
```

```python
all_data_task_embedding.shape
```

```python
import pandas as pd
import os
task_list = pd.read_csv(f'../data/{dataset_name}/task.csv', sep='\t')
all_data_task = pd.read_csv(f'../data/{dataset_name}/all_data_task.csv', sep='\t')
all_data = pd.read_csv(f'../data/{dataset_name}/all_data.csv', sep='\t')
task_occupation = pd.read_csv(f'../data/{dataset_name}/task_occupation.csv', sep='\t')
```

```python
from utils import cosine_similarity_matrix
similarity_matrix = cosine_similarity_matrix(all_data_task_embedding, task_occupation_embedding)
topk_indices = torch.topk(similarity_matrix, dim=-1,k=10).indices
topk_value = torch.topk(similarity_matrix, dim=-1,k=10).values
```

```python
import pandas as pd
topk_task = pd.DataFrame(topk_indices.numpy())
topk_values = pd.DataFrame(topk_value.numpy())
```

```python
for col in topk_task.columns:
    all_data_task[f"task{col}"] = topk_task[col].apply(lambda x: task_occupation['occupation'][x]) 
    all_data_task[f"task{col}_sim"] = topk_values[col] 
all_data_task.to_csv(f'outputs/{dataset_name}/all_data_task_topk.csv', index=None, sep='\t')
all_data_task.head()
```

```python
sample_k = 50
df = all_data_task.sort_values('task0_sim', ascending=False)[:sample_k].drop('text', axis=1)
df.to_excel(f'outputs/{dataset_name}/sample_top{sample_k}.xlsx')
df.head()
```
