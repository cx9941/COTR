```python
import pandas as pd
dataset_name = 'en'
df = pd.read_csv(f'../data/{dataset_name}/job_description.csv', sep='\t')
df
```

```python
import re

action_verbs = ["Prepare", "Organize", "Design", "Lead", "Complete", "Participate", "Managing", "Track", "Plan", "Communicate", "Manage"] + ["Manage", "Develop", "Plan", "Coordinate", "Implement", "Prepare", "Organize", "Oversee", "Lead", "Execute", "Support", "Monitor", "Receive", "Maintain"]
action_verbs = list(set(action_verbs))
keywords = ["develop", "prepare", "coordinate", "oversee", "lead", "support","is commited to"]

def extract_job_tasks(job_description):
    tasks = []

    if action_verbs:
        verb_pattern = r"|".join(action_verbs)
        delimiters = r'[\.!?\*\-]'
        task_pattern = re.compile(rf'\b(?:{verb_pattern})\b.{{4,}}?{delimiters}', re.DOTALL)
        tasks.extend(re.findall(task_pattern, job_description))
    
    if keywords:
        # 使用关键词列表进行匹配，捕获包含这些关键词的完整句子
        keyword_pattern = r"|".join(keywords)
        task_pattern = re.compile(rf'\b.{{2,}}?(?:{keyword_pattern})\b.{{4,}}?{delimiters}', re.DOTALL)
        tasks.extend(re.findall(task_pattern, job_description))
    tasks = list(set(tasks))
    return tasks

ans = extract_job_tasks(df['description'][0])
for idx, text in enumerate(ans):
    print(f"{idx}, {text}")
```

```python
from tqdm import tqdm
tqdm.pandas()
sub_df = df[:100]
sub_df['job'] = sub_df['description'].progress_apply(extract_job_tasks)
sub_df = sub_df.explode('job')
sub_df[~sub_df['job'].isna()]
```
