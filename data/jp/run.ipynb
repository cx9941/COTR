```python
import pandas as pd
import json
import pandas as pd
import os
ans = []
for i in os.listdir('raw_data'):
    if i.endswith('json'):
        temp = json.load(open(f'raw_data/{i}', 'r'))
        cols = ['Job_Title', 'Job_description']
        temp = pd.DataFrame(temp)[cols]
        temp.columns = ['job_title', 'job_description']
    if i.endswith('csv'):
        cols = ['title', 'description']
        temp = pd.read_csv(f'raw_data/{i}')[cols]
        temp.columns = ['job_title', 'job_description']
    ans.append(temp)
df = pd.concat(ans)
df.to_csv('job_description.csv', index=None, sep='\t')
df
```
