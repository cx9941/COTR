```python
import pandas as pd
import os

dataset_name = 'eu'
seed = 0
title_occupation = pd.read_parquet(f'../step4-title_occupation/results/{dataset_name}/seed{seed}.parquet')
job_task = pd.read_csv(f'../step2-task_map_occupation/outputs/{dataset_name}/title_task.csv', sep='\t')

occupation_task = pd.read_csv(f'../data/{dataset_name}/task_occupation.csv', sep='\t')
# occupation_task = occupation_task[~occupation_task['task'].isna()]
occupation_task = occupation_task.groupby('occupation').agg(list).reset_index()

title_occupation['title'] = title_occupation['title'].apply(lambda x: x.strip(' '))
job_task['title'] = job_task['title'].apply(lambda x: x.strip(' '))

title_occupation['occupation'] = title_occupation['occupation'].apply(lambda x: x.strip(' '))
occupation_task['occupation'] = occupation_task['occupation'].apply(lambda x: x.strip(' '))
occupation_task.columns = ['occupation', 'task_candidate']

job_task = pd.merge(job_task, title_occupation, on=['title'])
job_task = pd.merge(job_task, occupation_task, on=['occupation'])
if not os.path.exists(f'data/{dataset_name}'):
    os.makedirs(f'data/{dataset_name}')
job_task.to_parquet(f'data/{dataset_name}/task_candidate.parquet')
df = job_task[['description', 'task_candidate']]
df.columns = ['job description', 'task_candidate']
for seed in range(5):
    if not os.path.exists(f'data/{dataset_name}/seed{seed}/input0'):
        os.makedirs(f'data/{dataset_name}/seed{seed}/input0')
    df.to_parquet(f'data/{dataset_name}/seed{seed}/input0/task_candidate.parquet')
df
```
