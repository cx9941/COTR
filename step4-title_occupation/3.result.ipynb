```python
import pandas as pd
import os

dataset_name = 'en'
seed=1
result_dir = f'results/{dataset_name}/seed{seed}'
final_ans = []
task_topkoccupation = pd.read_parquet(f'data/{dataset_name}/task_topkoccupation.parquet')
task_topkoccupation.columns = ['origin_occupation', 'job description', 'occupation_candaite', 'sim']
task_topkoccupation = task_topkoccupation[['origin_occupation', 'job description', 'sim']]
task_topkoccupation.head()
```

```python
for i in range(len(os.listdir(result_dir))):
    df = pd.read_parquet(f"{result_dir}/result{i}.parquet")
    final_ans.append(df)
final_ans = pd.concat(final_ans)
final_ans = final_ans[['occupation', 'job description']].drop_duplicates('job description')
final_ans.to_excel(f"{result_dir}.xlsx", index=None)
final_ans.to_parquet(f"{result_dir}.parquet", index=None)
```

```python
final_ans = pd.merge(final_ans, task_topkoccupation, on=['job description']).sort_values(['sim', 'occupation'], ascending=[False, False])
topk = 100
final_ans[:topk].to_excel(f'results/{dataset_name}/seed{seed}_top{topk}.xlsx', index=None)
final_ans
```
