```python
import pandas as pd
import os

dataset_name = 'eu'
seed=0
result_dir = f'results/{dataset_name}/seed{seed}'
final_ans=[]
for i in range(len(os.listdir(result_dir))):
    df = pd.read_parquet(f"{result_dir}/result{i}.parquet")
    final_ans.append(df)
final_ans = pd.concat(final_ans)
final_ans = final_ans[['title', 'occupation']].drop_duplicates('title')
# final_ans.to_excel(f"{result_dir}.xlsx", index=None)
final_ans.to_parquet(f"{result_dir}.parquet", index=None)
final_ans
```

```python
task_topkoccupation = pd.read_parquet(f'data/{dataset_name}/seed{seed}/input0/occupation_candidate.parquet')[['title', 'occupation_sim']]
task_topkoccupation
```

```python
final_ans = pd.merge(final_ans, task_topkoccupation, on=['title']).sort_values(['occupation_sim', 'occupation'], ascending=[False, False])
topk = 100
# final_ans[:topk].to_excel(f'results/{dataset_name}/seed{seed}_top{topk}.xlsx', index=None)
final_ans
```
