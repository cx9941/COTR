```python
import os
import pandas as pd
dataset_name = 'en'
title_top10_occupation = pd.read_csv(f'../step3-bert/outputs/{dataset_name}/title_top10_occupation.csv', sep='\t')
title_top10_occupation['title'] = title_top10_occupation['title'].apply(lambda x: x.strip(' '))
title_top10_occupation['occupation_candidate'] = title_top10_occupation.apply(lambda x: [x[f"occupation{i}"] for i in range(10)], axis=1)
title_top10_occupation['occupation_sim'] = title_top10_occupation.apply(lambda x: sum([x[f"occupation{i}_sim"] for i in range(10)]) / 10, axis=1)
title_top10_occupation = title_top10_occupation[['title',  'occupation_candidate', 'occupation_sim']]
# for seed in range(5):
#     if not os.path.exists(f'data/{dataset_name}/seed{seed}/input0'):
#         os.makedirs(f'data/{dataset_name}/seed{seed}/input0')
#     title_top10_occupation.to_parquet(f'data/{dataset_name}/seed{seed}/input0/occupation_candidate.parquet')
title_top10_occupation
```

```python
title_top10_occupation = title_top10_occupation.sort_values('occupation_sim', ascending=False)
title_top10_occupation.to_excel(f'data/{dataset_name}/occupation_candidate.xlsx', index=None)
```
