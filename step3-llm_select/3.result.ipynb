```python
import pandas as pd
import os

dataset_name = 'en'
seed=0
result_dir = f'results/{dataset_name}/seed{seed}'
final_task = []
job_task = pd.read_csv(f'../step2-task_map_occupation/outputs/{dataset_name}/job_task.csv', sep='\t')[['title', 'task']]
job_task.columns = ['title', 'job description']
for turn in range(5):
    final_task.append(pd.read_parquet(f'results/{dataset_name}/seed{seed}/result{turn}.parquet')[['job description', 'task']])
final_task = pd.concat(final_task)
final_task = pd.merge(final_task, job_task, on=['job description'])
final_task
```

```python
occupation_candidate = pd.read_parquet(f'../step4-title_occupation/data/{dataset_name}/seed0/input0/occupation_candidate.parquet').drop_duplicates('title')
title_occupation = pd.read_parquet(f'../step4-title_occupation/results/{dataset_name}/seed0.parquet')
title_occupation = pd.merge(title_occupation, occupation_candidate[['title', 'occupation_sim']], on='title')
title_occupation
```

```python
final_task = pd.merge(final_task, title_occupation, on='title')[['title', 'job description', 'occupation', 'task', 'occupation_sim']]
final_task.columns = ['job title', 'job description', 'occupation', 'task', 'similarity']
final_task = final_task.sort_values('similarity', ascending=False)
final_task.to_csv(f'results/{dataset_name}/seed{seed}.csv', sep='\t', index=None)
# final_task.to_excel(f'results/{dataset_name}/seed{seed}.xlsx', index=None)
final_task
```
