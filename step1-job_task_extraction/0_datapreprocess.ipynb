```python
import os
import pandas as pd
dataset_name = 'jp'
df = pd.read_csv(f'../data/{dataset_name}/job_description.csv', sep='\t')
df = df[df['job_description'].apply(type)==type('')]
df = df[df['job_description'].apply(len)> 0]
df = df.drop_duplicates('job_description')
df.columns = ['job_title', 'job_description']
if not os.path.exists(f'data/{dataset_name}'):
    os.makedirs(f'data/{dataset_name}')
df.to_csv(f'data/{dataset_name}/job_description.csv', index=None, sep='\t')
df
```

```python
import re
 
# 示例字符串
text = "IP: 192.168.1.1 Port: 80"
 
# 正则表达式
pattern = ":"
 
# 使用正则表达式进行匹配
matches = re.findall(pattern, text)
 
# 输出匹配结果
for match in matches:
    print(match)
```
